```python
import pickle
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
folder='part-3/ct0141'
fmag=open(folder+'/ct0141-magnitude','rb')
fang=open(folder+'/ct0141-angle','rb')
fres=open(folder+'/ct0141-responses','rb')
```

```python
magnitude= pickle.load(fmag)
angle=pickle.load(fang)
responses=pickle.load(fres)
```

```python
def compute(mag,ang):
#mt={'data': image, 'label': 'opticflow'}
#savemat('optic_flow.mat',mt)
   X=np.linspace(0,127,128)
   Y=np.linspace(0,127,128)
  
   mg=mag.resize(128**2,1)
   ag=ang.resize(128**2,1)
   plt.figure()
   x,y=np.meshgrid(X,Y)

   dx=np.zeros(128**2)
   dy=np.zeros(128**2)
   for i in range(128**2):
       dx[i]=mag[i][0]*math.cos(ang[i][0])
       dy[i]=mag[i][0]*math.sin(ang[i][0])

   plt.quiver(x,y,dx,dy)

   plt.show()
```

```python
compute(magnitude[100],angle[100])
```
